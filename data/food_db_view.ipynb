```python
import pandas as pd

df = pd.read_csv("foundation_food.csv")

# drop duplicates and keep the first occurrence
df = df.drop_duplicates(subset=["description"], keep="first").dropna(subset=["diet_guide_group"]).reset_index()

food_nutrient_df = pd.read_csv("FoodData_Central_foundation_food_csv_2024-10-31/food_nutrient.csv")
nutrient_df = pd.read_csv("FoodData_Central_foundation_food_csv_2024-10-31/nutrient.csv").dropna().set_index("id")

results = []

for index, food in df.iterrows():
    result = food.to_dict()
    # result["nutrients"] = []
    food_nutrients = food_nutrient_df[food_nutrient_df["fdc_id"] == food["fdc_id"]][["nutrient_id", "amount"]]
    for index, row in food_nutrients.iterrows():
        # Get the nutrient_id and amount
        nutrient_id = row["nutrient_id"]
        amount = row["amount"]

        try:
            # Find the corresponding nutrient name
            nutrient = nutrient_df.loc[nutrient_id]
            result[nutrient["name"]] = amount
        except KeyError:
            # Handle the case where nutrient_id is not found
            print(f"Warning: Nutrient ID {nutrient_id} not found for food {food['description']}")
    # Append the result to the list
    results.append(result)

pd.DataFrame(results, index=None).to_csv("foundation_food_with_nutrients.csv", index=False)
```
